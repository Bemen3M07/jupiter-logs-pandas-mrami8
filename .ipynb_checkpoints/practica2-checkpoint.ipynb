```python
import logging
import logging.config
import time


logging.config.fileConfig('Logging.conf')


info_logger = logging.getLogger('infoLogger')
csv_logger = logging.getLogger('csvLogger')


while True:
   
    info_logger.info("Esto es un mensaje de información")
    info_logger.error("Esto es un mensaje de error")

  
    csv_logger.info("Mensaje de INFO registrado")

    time.sleep(2)


```

| **Mètode**                                             | **Exemple**                                                                                     | **Avantatges**                                                                                             | **Desavantatges**                                                                                          |
|--------------------------------------------------------|-------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------|
| **Configuració per defecte del mòdul `logging`**       | Utilitzar el mòdul `logging` amb la configuració per defecte. ## Exercici 5
**_Un cop tractades les dades introduïdes via fitxer caldrà investigar la llibreria que hagueu triat i elaborar com a mínim el codi necessari per presentar de forma que tinguin sentit amb l'àmbit del problema una sèrie de dades transformades obtingudes corresponents a, (s’haurà de adequar a les vostres dades però per a que tingueu algun exemple)._**

```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('datosej4.csv')

# Calcular la media de cada alumno
df['Mitjana'] = df[['M01', 'M04', 'M03', 'M05']].mean(axis=1)

# Calcular la media general de todas las notas
mitjana_general = df[['M01', 'M04', 'M03', 'M05']].mean().mean()
print(f"La mitjana general de tots els alumnes és: {mitjana_general:.2f}")

# Definir la nota mínima de aprobación
nota_aprovacio = 5

# Calcular el porcentaje de aprobados y suspensos
aprovats = df[df['Mitjana'] >= nota_aprovacio].shape[0]
suspesos = df[df['Mitjana'] < nota_aprovacio].shape[0]
total_alumnes = df.shape[0]
percentatge_aprovats = (aprovats / total_alumnes) * 100
percentatge_suspesos = (suspesos / total_alumnes) * 100
print(f"Percentatge d'aprovats: {percentatge_aprovats:.2f}%")
print(f"Percentatge de suspesos: {percentatge_suspesos:.2f}%")

# Calcular la nota més baixa i la més alta per a cada alumne
df['Nota_Més_Baixa'] = df[['M01', 'M04', 'M03', 'M05']].min(axis=1)
df['Nota_Més_Alta'] = df[['M01', 'M04', 'M03', 'M05']].max(axis=1)

# Calcular la desviación estándar de las notas de cada alumno
df['Desviacio'] = df[['M01', 'M04', 'M03', 'M05']].std(axis=1)

# Crear un resumen con las estadísticas
resum = df[['Professor', 'Alumne', 'Mitjana', 'Nota_Més_Baixa', 'Nota_Més_Alta', 'Desviacio']]
resum['Aprovacio'] = resum['Mitjana'].apply(lambda x: 'Aprovat' if x >= nota_aprovacio else 'Suspès')

# Encontrar la nota más baja y más alta a nivel global
nota_baixa_global = df[['M01', 'M04', 'M03', 'M05']].min().min()  # Nota mínima global
nota_alta_global = df[['M01', 'M04', 'M03', 'M05']].max().max()   # Nota máxima global

print(f"La nota més baixa a nivell global és: {nota_baixa_global}")
print(f"La nota més alta a nivell global és: {nota_alta_global}")

promedio_asignatura = df[['M01', 'M04', 'M03', 'M05']].mean()

# Imprimir el promedio de notas por asignatura sin el dtype
print("Promedio de notas por asignatura:")
print(promedio_asignatura.to_string())

```

## Exercici 6




```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('datosej4.csv')

# Calcular la media de cada alumno
df['Mitjana'] = df[['M01', 'M04', 'M03', 'M05']].mean(axis=1)

# Definir la nota mínima de aprobación
nota_aprovacio = 5

# Calcular el porcentaje de aprobados y suspensos
aprovats = df[df['Mitjana'] >= nota_aprovacio].shape[0]
suspesos = df[df['Mitjana'] < nota_aprovacio].shape[0]
total_alumnes = df.shape[0]
percentatge_aprovats = (aprovats / total_alumnes) * 100
percentatge_suspesos = (suspesos / total_alumnes) * 100

# Gráfico 1: Promedio de notas por asignatura (gráfico de barras)
promedio_asignatura = df[['M01', 'M04', 'M03', 'M05']].mean()
plt.figure(figsize=(8, 6))
promedio_asignatura.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red'])
plt.title('Promedio de notas por asignatura')
plt.ylabel('Promedio de notas')
plt.xlabel('Asignatura')
plt.xticks(rotation=0)
plt.show()

# Gráfico 2: Porcentaje de aprobados y suspensos (gráfico de pastel)
labels = ['Aprobados', 'Suspendidos']
sizes = [percentatge_aprovats, percentatge_suspesos]
colors = ['lightgreen', 'salmon']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Porcentaje de aprobados y suspensos')
plt.show()

# Gráfico 3: Nota más baja vs. Nota más alta (gráfico de dispersión)
df['Nota_Més_Baixa'] = df[['M01', 'M04', 'M03', 'M05']].min(axis=1)
df['Nota_Més_Alta'] = df[['M01', 'M04', 'M03', 'M05']].max(axis=1)
plt.figure(figsize=(8, 6))
plt.scatter(df['Nota_Més_Baixa'], df['Nota_Més_Alta'], color='purple')
plt.title('Nota más baja vs. Nota más alta')
plt.xlabel('Nota más baja')
plt.ylabel('Nota más alta')
plt.grid(True)
plt.show()

# Gráfico 4: Distribución de la media de las notas (histograma)
plt.figure(figsize=(8, 6))
df['Mitjana'].plot(kind='hist', bins=10, color='lightblue', edgecolor='black')
plt.title('Distribución de la media de las notas')
plt.xlabel('Media de notas')
plt.ylabel('Número de estudiantes')
plt.show()
```
